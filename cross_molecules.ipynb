```python
from rdkit import Chem
from lacan import lacan
from lacan import breed
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.drawOptions.drawMolsSameScale=False

# load parent molecules
m1 = Chem.MolFromSmiles("c1cc(ccc1[C@@H]2CCNC[C@H]2COc3ccc4c(c3)OCO4)F")
m2 = Chem.MolFromSmiles("CNCCC(C1=CC=CC=C1)OC2=CC=C(C=C2)C(F)(F)F")

# load lacan profile
p = lacan.load_profile("chembl")
# generate median molecules
median_molecules = breed.breed(m1,m2,p,nmols=9,cuts=2)
# display the molecules
d = Draw.MolsToGridImage(median_molecules,molsPerRow=3)
display(d)
```

```python
for m in median_molecules:
    print(Chem.MolToSmiles(m))
```
